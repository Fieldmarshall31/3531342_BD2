### **2.3.2:  Reading the CSV Data Files Using Pandas**

```python
products = pd.read_csv('products.csv')  # read the product csv files
reviews = pd.read_csv('reviews.csv')    # read the review csv files
users = pd.read_csv('users.csv')        # read the users csv files

# display the number of rows (records) and columns (attributes)
# in each data set
print("Products shape:\n", products.shape)
print("Reviews shape:\n", reviews.shape)
print("Users shape:\n", users.shape)
```

### **2.3.3:  Reading the Products.JSON File**

```python
# create an empty list to store each JSON objects
products_json = []

# Open the JSON file in read mode ('r')
for line in open('jcpenney_products.json', 'r'):
    # convert each line (string) into a dictionary
    products_json.append(json.loads(line))

# convert the appended list of dictionaries into a pandas DataFrame
# each dictionary becomes a row in the DataFrame
products_df = pd.DataFrame(products_json)

# display the number of rows (records) and columns (attributes)
# in the Products_df
print("Products_df shape:\n", products_df.shape)
```

### **2.3.4:  Reading the Reviewers.JSON File**

```python
# create an empty list to store each JSON objects
reviewers_json = []

# Open the JSON file in read mode ('r')
for line in open('jcpenney_reviewers.json', 'r'):
    # convert each line (string) into a python
    reviewers_json.append(json.loads(line))

# convert the appended list of dictionaries into a pandas DataFrame
# each dictionary becomes a row in the DataFrame
reviewers_df = pd.DataFrame(reviewers_json)

# display the number of rows (records) and columns (attributes)
# in the Reviewers_df
print("reviewers_df shape:\n", reviewers_df.shape)
```

# **Section 3:  Data Preparation**

## **3.1: Brief Overview of Section**

Following dataset inspection and selection, this section prepares the data for analysis by converting all the identifiers across the various data sets into lower case, creating a unified analytical dataset, and creating demographic variables (age and age groups) for segmentation.

## **3.2: Standardising the Column Names**

```python
# let's make the column names to lowercase to prevent
# an issue of case-sensitivity during merging
for df in [products, reviews, users, products_df, reviewers_df]:
    df.columns = df.columns.str.lower()
print(products.columns)
```

## **3.3: Merging the Selected Data Sets**

We would now go ahead to merge the **selected data sets** mentioned in **Section 2.8**, using the **reviews data set** as the the **central linking table**.

```python
# merging reviews with products
merged_ds = pd.merge(reviews, products, on='uniq_id', how='left')

# merging merged_ds with users
merged_ds = pd.merge(merged_ds, users, on='username', how='left')

print("Merged_ds shape:\n", merged_ds.shape)
```
