# **Section 3:  Data Preparation**

## **3.1: Brief Overview of Section**

Following dataset inspection and selection, this section prepares the data for analysis by converting all the identifiers across the various data sets into lower case, creating a unified analytical dataset, and creating demographic variables (age and age groups) for segmentation.

## **3.2: Standardising the Column Names**

```python
# convert column names to lowercase
for df in [products, reviews, users, products_df, reviewers_df]:
    df.columns = df.columns.str.lower()

print(products.columns)
print(reviews.columns)
print(users.columns)
```

## **3.3: Merging the Selected Data Sets**

```python
# merge reviews with products
merged_ds = pd.merge(reviews, products, on='uniq_id', how='left')

# merge merged_ds with users
merged_ds = pd.merge(merged_ds, users, on='username', how='left')
```

## **3.4: Creating the Demographic Variables**

### **3.4.1: Creating the Age Field**

```python
import datetime

# convert DOB to datetime format (dd.mm.yyyy)
merged_ds['dob'] = pd.to_datetime(merged_ds['dob'], errors='coerce', dayfirst=True)

# calculate the age
current_year = datetime.datetime.now().year
merged_ds['age'] = current_year - merged_ds['dob'].dt.year
```

### **3.4.2: Creating the Age-Group Field**

```python
bins = [0, 18, 25, 35, 50, 65, 100]
labels = ['<18', '18-25', '26-35', '36-50', '51-65', '65+']

merged_ds['age_group'] = pd.cut(merged_ds['age'], bins=bins, labels=labels)
```

# **Section 4:  Exploratory Data Analysis**

### **4.1: Brief Overview of this Section**

This section examines customer rating patterns and product performance.

I'll start with a quick analysis of the overall rating distribution to assess general satisfaction levels.

Next, I'll identify top-performing products based on average score rating and review score count. I'll also evaluate rating differences across age groups and states to understand demographic and geographic variations in customer satisfaction.

### **4.2: Distribution of Customer Rating Score**

```python
# frequency distribution table for review score count
merged_ds['score'].value_counts().sort_index()
```

```python
# bar chart showing how many times each rating score appears in the dataset
plt.figure()
sns.countplot(
    x='score', data=merged_ds,
    hue='score', edgecolor='black',
    palette='hls'
)

plt.title("Distribution of Customer Ratings", fontsize=12, weight= "bold")
plt.xlabel("Rating Score", fontsize=10, weight= "bold")
plt.ylabel("Frequency", fontsize=10, weight= "bold")

plt.tight_layout()
plt.show()
```

This chart displays the visual representation of the frequency distribution count of the product reviews score.

From this chart, we can derive that the review score is higher between 0 and 1 with 4 and 5 having lower frequency.

### **4.3: Product Performance Analysis for Top 10 Rated Products**

```python
# average rating and total reviews for each product
product_perf = (
    merged_ds.groupby('name')
    .agg(
        average_score=('score', 'mean'),  # mean score per product
        review_count=('score', 'count')   # total score count per product
    )
    .reset_index()
)

# products with at least 10 reviews
product_perf = product_perf[product_perf['review_count'] >= 10]

# top 10 highest-rated products
top_10_products = product_perf.sort_values(
    'average_score',
    ascending=False
).head(10)
```

```python
top_10_products.head(10)
```

```python
# horizontal bar chart showing
# the top 10 highest-rated products based on average score
plt.figure(figsize=(10,6))

sns.barplot(
    x='average_score',
    y='name',
    data=top_10_products,
    edgecolor='black',
    hue='name',
    palette='magma'
)

plt.title("Top 10 Performing Products (Min. 10 Reviews)", fontsize=14, weight= "bold")
plt.xlabel("Average Rating", fontsize=12, weight= "bold")
plt.ylabel("Product Name", fontsize=12, weight= "bold")

plt.tight_layout()
plt.show()
```

The chart shows the ten products with the highest average rating score, considering only those with a minimum of 10 review scores to ensure reliable results. The GE® ENERGY STAR® 4.3 DOE Cu. ft. High Efficiency Capacity Front-Load Washer has the highest average ratings.

However, the average ratings of the top products remain moderate, generally ranging between about 2.3 and 3.1. This aligns with the earlier rating distribution analysis, which showed that lower ratings dominate the dataset.

### **4.4: Review Rating Score Distribution for Top 10 Rated Products**

```python
# dataset containing only the top 10 products
top_products_scores = merged_ds[merged_ds['name'].isin(top_10_products['name'])]

# order products by average score (highest first)
order = top_10_products.sort_values('average_score', ascending=False)['name']
```

```python
# boxplot showing rating distribution for each top product
plt.figure(figsize=(10,6))

sns.boxplot(
    x='score',
    y='name',
    data=top_products_scores,
    order=order,  # ensures highest average score appears first
    palette='Set2',
    hue='name'
)

plt.title("Review Rating Score Distribution for Top 10 Products", fontsize=12, weight='bold')
plt.xlabel("Review Rating Score", fontsize=10, weight= "bold")
plt.ylabel("Product Name", fontsize=10, weight= "bold")

plt
plt.tight_layout()
plt.show()
```

This chart shows the distribution of customer review score for the ten products identified as the top reviewed items in the previous analysis. Overall, ratings for these products are spread across several score levels, indicating that customer review score is varied rather than consistently high.

The GE® ENERGY STAR® 4.3 DOE Cu. ft. High Efficiency Capacity Front-Load Washer, which recorded the highest average rating, also shows ratings across multiple values. A similar pattern is observed for other products such as Xersion™ Compression Slider Shorts and Call It Spring™ Chaella Ballet Flats, where ratings vary across the scale.

This variation explains why the average ratings fall within a moderate range of approximately 2.3 to 3.1.

### **4.5: Average Rating by Age Group**

```python
# average rating score per age group
age_perf = (
    merged_ds.groupby('age_group', observed=False)['score']
    .mean()
    .reset_index()
)
```

```python
age_perf[['age_group', 'score']].head(6)
```

```python
# bar chart showing the average rating score per age group
plt.figure(figsize=(8,5))

sns.barplot(
    x='age_group',
    y='score',
    data=age_perf,
    edgecolor='black',
    hue='age_group',
    palette='flare'
)

plt.title("Average Rating by Age Group", fontsize=12, weight= "bold")
plt.xlabel("Age Group", fontsize=10, weight= "bold")
plt.ylabel("Average Rating", fontsize=10, weight= "bold")

plt.tight_layout()
plt.show()
```

The chart shows the average product rating across different age groups, with ratings remaining relatively similar for all categories, generally ranging between about 1.4 and 1.5. The 36-50 age group records the slightly highest average rating score, followed closely by the 51-65 and 65+ groups. The 18-25 age group gives the lowest average rating score.

### **4.6: Top 10 States by Average Rating**

```python
# average rating score per state
state_perf = (
    merged_ds.groupby('state')['score']
    .mean()
    .reset_index()
)

# top ten states with the highest average rating score
top_ten_states = state_perf.sort_values(
    'score',
    ascending=False
).head(10)
```

```python
top_ten_states[['state', 'score']].head(10)
```

```python
# horizontal bar chart showing the top 10 states
# with highest average rating score
sns.barplot(
    x='score',
    y='state',
    data=top_ten_states,
    hue = 'state',
    edgecolor='black',
    palette='mako'
)

plt.title("Top 10 States by Average Rating", fontsize=12, weight= "bold")
plt.xlabel("Average Rating", fontsize=10, weight= "bold")
plt.ylabel("Top 10 States", fontsize=10, weight= "bold")

plt.tight_layout()
plt.show()
```

The chart displays the ten states with the highest average customer ratings score, with Indiana recording the highest score, followed by Georgia, the U.S. Virgin Islands, and Florida. The remaining states also show similar ratings, with averages clustered between approximately 1.45 and 1.60.
